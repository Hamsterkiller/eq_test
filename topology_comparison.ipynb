```python
import pandas as pd
```

```python
node_matlab = pd.read_csv("./test_data/node.csv", sep=",")
node_matlab.drop(['idx'], inplace=True, axis=1)
node_cpp = pd.read_csv("./test_data/bus_in.csv", sep=",", header=None)
line_matlab = pd.read_csv("./test_data/vetv.csv", sep=",")
line_matlab.drop(['idx'], inplace=True, axis=1)
line_cpp = pd.read_csv("./test_data/line_in.csv", sep=",", header=None)
```

```python
node_cpp.columns = ["node", "type", "vzd", "unom", "pg", "pn", "qg", "qn", "gsh", "bsh"]
node_cpp = node_cpp[node_matlab.columns]
node_cpp.pg = node_cpp.pg * 100
node_cpp.pn = node_cpp.pn * 100
node_cpp.qg = node_cpp.qg * 100
node_cpp.qn = node_cpp.qn * 100
node_matlab.qn = node_matlab.qn * 100
line_cpp.columns = ["node_from", "node_to", "flow", "pnum_actual", "pnum", "type", 
                    "r", "x", "g", "b", "b_from", "b_to", "ktr", "kti", "ts_real",
                    "ts_imag", "node_from_orig", "node_to_orig"]
```
